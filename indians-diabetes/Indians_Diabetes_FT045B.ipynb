```python
# Média de idade para cada grupo (diabéticos e não-diabéticos)
diabetes_df.groupby('Diabetes')['Idade'].mean()
```

```python
# Várias métricas por grupo
diabetes_df.groupby('Diabetes').agg({
    'Idade': ['mean', 'max'],
    'IMC': ['mean', 'min']
}).reset_index() # A função reset_index serve para ajustar avisualização.
```

```python
# Contagem de Ocorrências Únicas
diabetes_df['Gravidez' ].value_counts()
```

```python
# Pacientes com mais de 30 anos
diabetes_df[diabetes_df['Idade'] > 30]
```

```python
# Pacientes com mais de 30 anos E IMC acima de 25
diabetes_df[(diabetes_df['Idade'] > 30) & (diabetes_df['IMC'] >25)]
```

```python
# Pacientes com diabetes OU IMC acima de 35
diabetes_df[(diabetes_df['Diabetes'] == 1) | (diabetes_df['IMC']> 35)]
```

```python
# Ordenar por idade (crescente)
diabetes_df.sort_values('Idade')

# Ordenar por IMC (decrescente)
diabetes_df.sort_values('IMC', ascending=False)

# Ordenar por múltiplos critérios
diabetes_df.sort_values(by=['Diabetes', 'Idade'], ascending=[False, True])
```

```python
# Ordenar por IMC (decrescente)
diabetes_df.sort_values('IMC', ascending=False)
```

```python
# Ordenar por múltiplos critérios
diabetes_df.sort_values(by=['Diabetes', 'Idade'], ascending=[False, True])
```

```python
# Histograma da idade
plt.figure(figsize=(8, 6))
sns.histplot(data=diabetes_df, x='Idade', kde=True, hue='Diabetes')
plt.title('Distribuição da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Scatter plot Glicotese vs. IMC, colorido por Diabetes
plt.figure(figsize=(10, 8))
sns.scatterplot(data=diabetes_df, x='Glicotese', y='IMC', hue='Diabetes', alpha=0.6)
plt.title('Glicotese vs. IMC por Diabetes')
plt.xlabel('Glicotese')
plt.ylabel('IMC')
plt.show()
```

```python
# Box plot de IMC por Diabetes
plt.figure(figsize=(8, 6))
sns.boxplot(data=diabetes_df, x='Diabetes', y='IMC')
plt.title('Box Plot de IMC por Diabetes')
plt.xlabel('Diabetes (0: Não, 1: Sim)')
plt.ylabel('IMC')
plt.show()
```

```python
# Contagem de casos de Diabetes
plt.figure(figsize=(6, 4))
sns.countplot(data=diabetes_df, x='Diabetes')
plt.title('Contagem de Casos de Diabetes')
plt.xlabel('Diabetes (0: Não, 1: Sim)')
plt.ylabel('Contagem')
plt.show()
```
